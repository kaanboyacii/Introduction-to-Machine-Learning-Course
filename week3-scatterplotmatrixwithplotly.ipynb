```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from plotly.offline import iplot
import plotly.graph_objs as go
import plotly.figure_factory as ff
```

```python
timesData = pd.read_csv("/kaggle/input/university-rankings-2011/University_Rankings_20220204.csv")
```

```python
timesData.head()
```

```python
dataframe = timesData[timesData.year == 2014]
```

```python
data2014 = dataframe.loc[:,["research","citations", "teaching"]] 
```

```python
data2014["index"] = np.arange(1,len(data2014)+1) 
```

```python
data2014.head()
```

```python
fig = ff.create_scatterplotmatrix( 
                                  data2014, 
                                  diag='box',  
                                  index='index',
                                  colormap='Portland', 
                                  colormap_type='cat',  
                                  height=700, width=700)
iplot(fig)
```
