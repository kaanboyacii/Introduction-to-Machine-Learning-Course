```python
import numpy as np
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.arange(0, 20, 0.2)
y = np.sin(x)
plt.plot(x, y)
plt.title("sinüs ")
plt.xlabel("X ekseni")
plt.ylabel("Y ekseni")
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = np.random.randint(75, 250, 50)
plt.scatter(x, y, c=colors, s=sizes)
plt.title("Point Chart")
plt.xlabel("X-eks")
plt.ylabel("Y-eks")

plt.show()
```

```python
data = np.random.randn(750)
plt.hist(data, bins=75)
plt.title("hist")
plt.xlabel("deger")
plt.ylabel("frekans")
plt.show()
```

```python
x = ['A', 'B', 'C', 'D', 'E']
y = np.random.randint(5, 16, 5)
plt.bar(x, y)
plt.title("chart")
plt.xlabel("kategori")
plt.ylabel("deger")

plt.show()
```

```python
sizes = [5, 25, 20, 10, 5, 25]

plt.pie(sizes)

plt.title("pie chart")

plt.show()
```

```python
x = np.arange(-10, 10, 0.10)
y = np.arange(-10, 10, 0.10)
x, y = np.meshgrid(x, y)
r = np.sqrt(x**2 + y**2)
z = np.sin(r)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z)
ax.set_title("3D Graphics")
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z axis")
plt.show()
```
